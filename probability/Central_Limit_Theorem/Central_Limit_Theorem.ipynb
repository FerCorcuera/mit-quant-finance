## Exercise 1:

- Creating a random variable with a binomial distribution and see how it changes as K grows

```python
n = 20  # number of trials
p = 0.1  # probability of success
k = np.arange(0, n + 1)  #

b = binom.pmf(k, n, p)

plt.bar(k, b)
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title(f"Binomial distribution of k successes with {p} of succes in {n} trials")

theoretical_mean = n * p
sample_mean = np.sum(k * b)

print(f"Theoretical mean:{theoretical_mean} \n vs \nSample mean: {sample_mean}")
```

## Exercise 2:

- Standardazing the binomial random variable

```python
n = 20  # number of trials
p = 0.1  # probability of success
k = np.arange(0, n + 1)  #

b = binom.pmf(k, n, p)

standard_dev = sqrt(n * p * (1 - p))
mean = n * p

z = (k - mean) / standard_dev

zmax = 5
zi = np.abs(z) <= 5

plt.bar(z[zi], b[zi])
plt.xlabel("Standardized outcome (z)")
plt.ylabel("Probability")
plt.title(f"Binomial distribution of k successes with {p} of succes in {n} trials")
```

## Exercise 3: Bloomberg terminals

- We have 120 students
- Only 9 bloomberg terminals
- We define that each student wants or not to use the terminal
- With a probability of 7.5%
- What is the probability of 9 or less students usign the terminals at teh same time?

```python
p = 0.075
n = 120
k = np.arange(0, 10)

b = binom.pmf(k, n, p)

answer = sum(b)

print(f"The probability of 9 or less students using the terminals is: {answer:.1%}")
```
