# Monte Carlo simulation

> Content from the MITx course 'Mathematical Methods for quant finance, unit 2'

- It is a tool to simulate probability distributions, experiments or stochastic processes using computer power
- It is limited to the computed features (concept of random, range of numbers)
- It is a great tool to what if questions, approximate answers and understand financial series, simulating reality many times trying to have approximate answers empirically

```python
import numpy as np
from scipy.stats import binom
from matplotlib import pyplot as plt
from math import sqrt
import seaborn as sns
```

## 1. 20 step random walk simulation

```python
unif = np.random.default_rng()
p = 0.5
q = 1 - p
Nt = 20
Np = 1

z = unif.random((Nt, Np))
x = np.sign(p - z)
s = np.zeros((Nt + 1, Np))

for i in range(1, len(s)):
    s[i] = s[i - 1] + x[i - 1]


plt.figure(figsize=(8, 4))
plt.plot(s, marker="o")
plt.title("Recursive random walk simulation of 20 periods - 1 path")
plt.grid()
plt.show()
```

## Many simulations of a one-year daily walk

```python
unif = np.random.default_rng()
p = 0.5
q = 1 - p
Nt = 252  # trading days in a year
Np = int(1e4)  # 10,000 simulations

z = unif.random((Nt, Np))
x = np.sign(p - z)
s = np.zeros((Nt + 1, Np))

for i in range(1, len(s)):
    s[i] = s[i - 1] + x[i - 1]  # since we are using numpy, this affects all i-row

plt.figure(figsize=(8, 4))
plt.plot(s[:, 0:3], marker=".")
plt.title("Recursive random walk simulation of 20 periods - 1 path")
plt.grid()
plt.show()
```
