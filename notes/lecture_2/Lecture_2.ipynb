```python
import numpy as np
```


## Task 1: Represent the economy
 - A0 → vector of prices at t=0
 - A → matrix of payoffs at T (scenarios × assets)


```python
A0 = np.array([[100,100]])
A = np.array([[103,90],[103,130]])
```


## Task 2: Portfolio valuation
Implement:
$V_0(\pi) = A_0 \cdot \pi$
$V_T(\pi) = A \pi$

Matrix–vector multiplication requires matching dimensions because a portfolio payoff is a linear combination of asset payoff vectors, with one coefficient per asset.



```python
pi = np.array([[10],[10]])
V0 = A0 @ pi
V1 = A @ pi
```

### Task 3: Replicating portfolio
Solve:
$A \pi = C_T$

Using `np.linalg.solve`:

Given that our contingent claim in this scenario is a put option with strike price K = 100, so when the state of the market is down (d) , the payoff would be the stirke price minus the down price, and when the market is up, the payoff would be 0 (we would prefer to sell our asset in the market rather that using our put option)


```python
# given that the first row is the down state, we write our contingent claim with two rows , the first for the down and the second 
# for the up state
CT = np.array([[10],[0]])
pi_star = np.linalg.solve(A,CT)
pi_star
# pi star has the portfolio wieghts that replicate the payoffs of the contingent claim
# this only works when assets are linearly independent (one solution)
# market is complete
```

## Task 4: Arbitrage detection
Check if there exists:
$\pi \quad \text{s.t.} \quad V_0(\pi) \le 0,\; V_T(\pi) \ge 0,\; \exists \omega: V_T(\pi) > 0$


```python
from itertools import product
```

```python
test = [[-10,-1,1,10],[-10,-1,1,10]]
weights = list(product(*test))
weights
```

```python
arbitrage_portfolio = []
for x in range(len(weights)):
    pi_test = np.array([[weights[x][0]],[weights[x][1]]])

    V1 = A @ pi_test

    V0 = A0 @ pi_test

    if V0 <= 0 and V1[0] >= 0 and V1[1] >= 0 and any(x > 0 for x in V1):
        arbitrage_portfolio.append(pi_test)

arbitrage_portfolio
# there is not any arbitrage portfolio in the set of portfolios that we've checked
```

## 5. Using simplex to check if the inequality exisits:

```python
# first we define our objective vector
c = [0,0]

# then out inequality matrix
A_ub = [
    [100,100],
    [-103,-90],
    [-103,-130]
]

b_ub = [0,0,0]

from scipy.optimize import linprog

res = linprog(c, A_ub = A_ub, b_ub=b_ub, method = "simplex")
res
```
